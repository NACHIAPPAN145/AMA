```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_excel('/content/AMA GROUP.xlsx')
print(df.head())
```

```python
# Initialize empty list to store the rearranged data
rearranged_data = []
# Loop over each row of the original data
for index, row in df.iterrows():
    register_num = row['RegNum']
    choice = row['Choice']

    # Loop through each profile (Profile1 to Profile13)
    for i in range(1, 14):
        profile_id = f'Profile{i}'
        rating = row[profile_id]

        # Check if the profile is the selected choice
        is_choice = 1 if choice == f'Profile{i}' else 0

        # Add a new row to the rearranged data
        rearranged_data.append([register_num, i, rating, is_choice])
# Create a DataFrame from the rearranged data
rearranged_df = pd.DataFrame(rearranged_data, columns=['RegisterN', 'ProfileID','Rating', 'Choice'])
rearranged_df.head(15)
```

```python
# prompt: download the new dataframe

from google.colab import files
rearranged_df.to_csv('rearranged_data.csv', encoding = 'utf-8-sig')
files.download('rearranged_data.csv')
```
